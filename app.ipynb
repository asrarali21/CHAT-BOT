```python
from langchain_community.document_loaders import WebBaseLoader
from langchain_community.vectorstores import Qdrant
from langchain_qdrant import Qdrant
from qdrant_client import QdrantClient
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_google_genai import ChatGoogleGenerativeAI




loader = WebBaseLoader("https://www.chitika.com/scaling-rag-20-million-documents/")
docs = loader.load()
docs

docs_spilt = RecursiveCharacterTextSplitter(  
    chunk_size=800, 
    chunk_overlap=150
    )

documents = docs_spilt.split_documents(docs)

docs_embedding = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")

client = QdrantClient(
    "localhost", port=6333
)

vectorStore = Qdrant.from_documents(
    documents=documents,
    embedding=docs_embedding

)
```
